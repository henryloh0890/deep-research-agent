```python
from dotenv import load_dotenv
import os

load_dotenv("../.env")

print("Anthropic key loaded:", bool(os.getenv("ANTHROPIC_API_KEY")))
print("Groq key loaded:", bool(os.getenv("GROQ_API_KEY")))
```

```python
from langchain_anthropic import ChatAnthropic
from langchain_openai import ChatOpenAI

# Primary LLM - Claude (best for writing & synthesis)
claude = ChatAnthropic(model="claude-sonnet-4-5", temperature=0)

# Fast & free LLM - Groq/Llama
groq_llm = ChatOpenAI(
    model="llama-3.3-70b-versatile",
    base_url="https://api.groq.com/openai/v1",
    api_key=os.getenv("GROQ_API_KEY"),
    temperature=0
)

# Sanity check both
print("Claude:", claude.invoke("Say hello in one sentence.").content)
print("Groq:", groq_llm.invoke("Say hello in one sentence.").content)
```

```python
from langchain_community.tools import DuckDuckGoSearchResults

search = DuckDuckGoSearchResults(num_results=5)

# Test the search
results = search.invoke("AI agent frameworks 2025")
print(results)
```

```python
import warnings
warnings.filterwarnings("ignore")

from langgraph.prebuilt import create_react_agent
from langchain_core.messages import HumanMessage

# Create agent with search tool attached
agent = create_react_agent(claude, tools=[search])

# Test run the agent
result = agent.invoke({
    "messages": [HumanMessage(content="Research the latest trends in AI agents in 2025. Give me 5 key insights.")]
})

# Print the final response
print(result["messages"][-1].content)
```

```python
# Showing every step the agent took — search queries, results, reasoning
for message in result["messages"]:
    role = message.__class__.__name__
    content = message.content if isinstance(message.content, str) else str(message.content)
    print(f"[{role}]")
    print(content[:500])
    print("---")
```
